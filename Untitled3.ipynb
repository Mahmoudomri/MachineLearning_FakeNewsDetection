```python
import pandas as pd
import numpy as np


data2= pd.read_csv('/content/Fake.csv')
data1=pd.read_csv('/True.csv')

data=pd.concat([data1,data2]).reset_index(drop=True)
```

## **Data cleaning**

```python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt')
from nltk.tokenize import word_tokenize

from nltk.stem import WordNetLemmatizer   #transform the word to it basic
lemmatizer=WordNetLemmatizer()

from nltk.stem.snowball import SnowballStemmer
stemmer=SnowballStemmer(language="english")

from nltk.stem.porter import PorterStemmer
stemmer2=PorterStemmer()

from nltk.stem.lancaster import LancasterStemmer
stemmer3=LancasterStemmer()

from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))

import string
punct =string.punctuation
```

```python
print(lemmatizer.lemmatize("indoctrination"))
print(stemmer.stem("indoctrination"))
print(stemmer2.stem("indoctrination"))
print(stemmer3.stem("indoctrination"))

print("/n")

print(lemmatizer.lemmatize("sophisticating"))
print(stemmer.stem("sophisticating"))
print(stemmer2.stem("sophisticating"))
print(stemmer3.stem("sophisticating"))

print("/n")

print(lemmatizer.lemmatize("appeles"))
print(stemmer.stem("appeles"))
print(stemmer2.stem("appeles"))
print(stemmer3.stem("appeles"))
```

```python
list(stop_words)[:10]
```

```python
punct
```

```python
def process (s) :
  for p in punct:
    s=s.replace(p,'')
  s=s.lower()
  s=word_tokenize(s)
  s=[w for w in s if not w in stop_words]
  s=[stemmer.stem(word) for word in s]
  return s
```

```python
process("apples im Mahmoud!")
```

```python
number_of_samples =data.shape[0]
from tqdm import tqdm
for i in tqdm(range(number_of_samples)):
  data[i]=process(data[i])
```

# ***Modelling***

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.naive_bayes import MultinomialNB
nb=MultinomialNB()
```

```python
def train_and_evaluate(x,y):
  xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
  nb.fit(xtrain,ytrain)
  ypred_tr=nb.predict(xtrain)
  ypred_ts=nb.predict(xtest)
  print("training resulats :\n")
  print(classification_report(ytrain,ypred_tr))
  print("test resulats :\n")
  print(classification_report(ytest,ypred_ts))
```

```python
train_and_evaluate(bow_data,y)
```
